```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import datetime
```

```python
df_set=df[['DAY_OF_MONTH','DAY_OF_WEEK','TAIL_NUM','ORIGIN','DEST','DEP_DEL15','DEP_TIME_BLK','ARR_DEL15','DISTANCE']]
df_set.columns=df_set.columns.str.lower()
df_set
```

### DATA CLEANSING

```python
def missing_value():
    mv=df_set.isnull().sum().reset_index()
    mv.columns=['fiture','total_missing_value']
    mv['%']=round(mv['total_missing_value']/len(df_set)*100,2)
    mv=mv[mv['%']>0].sort_values('%',ascending=False)
    fitur=mv['fiture']
    return mv
mv=missing_value()
mv
```

```python
fitur=mv['fiture'].tolist()
for i in range(0,len(fitur)):
    df_set=df_set.dropna(subset=[fitur[i]])
missing_value()
print(len(df_set))
```

```python
df_set['dep_time_blk'].unique()
```

```python
df_set['dep_time']=(pd.to_numeric(df_set['dep_time_blk'].str.split(pat='-',expand=True)[0])/100).astype(int)
df_set=df_set.sort_values(['day_of_month','day_of_week','tail_num','dep_time'], ascending=True).reset_index(drop=True)
df_set
```

```python
tn0=''
ad0=0
dep0=0
tn1=''
ad1=0
dep1=0
s=[]
for i, k in df_set.iterrows():
    tn1=k['tail_num']
    ad1=k['arr_del15']
    dep1=k['dep_time']
    if tn1 == tn0:
        if dep1 > dep0:
            if ad0 !=0:
                temp=1
                s.append(temp)
            else: 
                temp=0
                s.append(temp)     
        else:
            temp=0
            s.append(temp)
    else:
        temp=0
        s.append(temp)
        
    tn0=tn1
    ad0=ad1
    dep0=dep1
df_set['prev._arr_del']=s
df_set
```

### Kalau ini aku grouping tail_num berdasarkan tanggal, trs nampilin delay dan not delay berdasarkan jumlah penerbangan pada tanggal itu

```python
dmonth=pd.DataFrame(df_set.groupby(["tail_num","dep_del15","day_of_month"])["dep_del15"].count().unstack("dep_del15").reset_index().values,columns=["Tailnum","Tgl","Not_Delayed","Delayed"])
dmonth["Percent Of Delayed"]=dmonth.Delayed /( dmonth.Delayed+ dmonth.Not_Delayed)
dmonth.sort_values(by="Tailnum",ascending=False,inplace=True)
dmonth.head()
```

```python
dmerged = dl.merge(dmonth, left_on=['tail_num','day_of_month'],
              right_on=['Tailnum','Tgl'],
              how='left')
dmerged.sort_values(by="tail_num",ascending=False,inplace=True)
dmerged.head()
```

```python
dmerged=dmerged.drop_duplicates(subset=['day_of_month','day_of_week','tail_num','origin','dest','dep_time',
                                        'Tailnum','Not_Delayed'])
len(dmerged)
dmerged.sort_values(by="tail_num",ascending=True,inplace=True)
dmerged.head(10)
```
